```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,3,100) #linspace makes an array 0-3 with 100 data points
yerr=0.05+2*x #y-error for error bars

plt.figure()
plt.title('exponential plots')
plt.xlabel('x values')
plt.ylabel('exponents')

#plt.plot(x,x**2,label='x^2')
#plt.plot(x,x**3,label='x^3')
#plt.plot(x,x**4,label='x^4')
#plt.plot(x,x**5,label='x^5')

for n in range (2,6): #an easier way to do the above is making a for loop
    #plt.plot(x,x**n, label='x^'+str(n))
    plt.errorbar(x,x**n,yerr=yerr, label='x^'+str(n)) #errorbar is a plot in itself, so either use plt.plot or plt.errorbar with error bars

plt.legend() #creates a legend based on the labels
plt.show()
```
