```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset('planets')
#data2=pd.DataFrame.dropna(data) doing this makes it all sad looking?
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
data
```

```python
#first plot
plot=sns.scatterplot(data=data,x='orbital_period',y='mass',hue='method',palette='Paired')
plt.xscale('log')
plt.yscale('log') #do both?
plot.set(xlabel='orbital period',ylabel='mass',title='p-mass relationship')

#https://seaborn.pydata.org/generated/seaborn.scatterplot.html
#https://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
#2nd plot
data2=pd.DataFrame.dropna(data)

plot2=sns.barplot(data=data2,x="year",y="number",hue="method",palette="Paired")
plot2.set(ylabel='planets discovered',title='planets discovered per year') 
plot2.set_xticklabels(plot2.get_xticklabels(), rotation=60) #https://stackoverflow.com/questions/66845482/seaborn-box-plot-x-axis-too-crowded

#the graph is kind of ugly... i don't know why the colors are like that??

#also, I got a warning after running:
##C:\Users\roost\AppData\Local\Temp\ipykernel_34912\4077675205.py:4: UserWarning: set_ticklabels() should only be used with a fixed number of ticks, i.e. after set_ticks() or using a FixedLocator.
  #plot2.set_xticklabels(plot2.get_xticklabels(), rotation=60)
```
